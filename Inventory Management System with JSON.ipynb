```python
#importing the libraries
import json
from datetime import datetime

# Read the existing inventory from the JSON file
fd = open("Records.json", "r")
js = fd.read()
fd.close()

record = json.loads(js)# Parse the JSON string into a Python dictionary

print("-----------------------MENU---------------------\n")
for key in record.keys():
    print(key, ":" ,record[key]["Name"] ,"\t|",record[key]["Price"],"|", record[key]["Qn"])
print("--------------------------------------------------")

#input from user
ui_name =  str(input("Enter your Name        : "))
ui_mail =  str(input("Enter your Email id    : "))
ui_phone = str(input("Enter your Phone No.   : "))
ui_pr =    str(input("Enter a Product id :"))
ui_qn =    int(input("Enter a product Qn : "))

# GST rate (assuming 18%)
gst_rate = 0.18

# If we have enough or more quantity then user wants 
if(record[ui_pr]["Qn"] >= ui_qn ):
    
    #Updating our inventory File 
    record[ui_pr]["Qn"] = record[ui_pr]["Qn"]-ui_qn
    print("--------------------Billing Amount ------------------------\n")
    print("Name of the Product  : " ,       record[ui_pr]["Name"])
    print("Price of Product     : ",  "Rs",record[ui_pr]["Price"])
    print("Your Quantuty        : " ,       ui_qn)
    print("-----------------------------------------------------------------")
    print("Biiling Amount       : " ,"Rs",record[ui_pr]["Price"] * ui_qn)
    print("\n")

      # Calculate GST
    gst_amount = record[ui_pr]["Price"] * gst_rate
    total_amount = (record[ui_pr]["Price"] * ui_qn) + (gst_amount * ui_qn)
    
    print("GST (18%)                            : ", "Rs", gst_amount * ui_qn)
    print("Total Billing Amount after GST       : ", "Rs", total_amount)
    print("\n")
    print("--------------------Inventory Updated--------------------\n")
    
    # Prepare the sale record
    sale =  {
        "Product ID": ui_pr,
        "Product Name": record[ui_pr]["Name"],
        "Quantity": ui_qn,
        "Price per Unit": record[ui_pr]["Price"],
        "Billing Amount": record[ui_pr]["Price"] * ui_qn,
        "Customer Name": ui_name,
        "Customer Email": ui_mail,
        "Customer Phone": ui_phone,
        "Date & Time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
}
    
# If we have  not enough quantity then user wants
else:
    #Dispaly Informaion about Inventory we have 
    
    print("Sorry we do not have much quantity")
    print(f"We have only {record[ui_pr]["Qn"]}  quantity")
    print("Would you like to purchase them?")
    print("--------------------------------------------------")
    

    ch = str(input("if yu want Press Y/y :"))
    
    if( ch == "Y" or ch == "y"):
        
        print("--------------------Billing Amount ------------------------/n")
        
        print("Name of the Product  : " ,  record[ui_pr]["Name"])
        print("Price of Product     : ",  "Rs",record[ui_pr]["Price"])
        print("Your Quantuty        : " ,  record[ui_pr]["Qn"])
        print("--------------------------------------------------")
        print("Biiling Amount       : " ,"Rs",record[ui_pr]["Price"] * record[ui_pr]["Qn"] )

        
         # Calculate GST
        gst_amount = record[ui_pr]["Price"] * gst_rate
        total_amount = (record[ui_pr]["Price"] * record[ui_pr]["Qn"]) + (gst_amount * record[ui_pr]["Qn"])
        
        print("GST (18%)                           : ", "Rs", gst_amount * record[ui_pr]["Qn"])
        print("Total Billing Amount after GST      : ", "Rs", total_amount)
        
        # Prepare the sale record
        sale = {
            "Product ID": ui_pr,
            "Product Name": record[ui_pr]["Name"],
            "Quantity": record[ui_pr]["Qn"],
            "Price per Unit": record[ui_pr]["Price"],
            "Billing Amount": record[ui_pr]["Price"] * record[ui_pr]["Qn"],
            "Customer Name": ui_name,
            "Customer Email": ui_mail,
            "Customer Phone": ui_phone,
            "Date & Time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            
            
        }
        
        #Updating our inventory File
        record[ui_pr]["Qn"] = 0
        
    else:
        
        #if user press anything other then Y/y 
        print('Purchase cancel')
        sale = None

        
    
        
    

print("Thanku For ur ResponseðŸ˜Š")






js = json.dumps(record) # Convert back  dictionary into JSON string

# Append the sale to SalesLog.json
fd = open("Sales.txt","a")
fd.write(json.dumps(sale) + "\n")
fd.close()

# Save updated inventory to file
fd = open("Records.json","w")
fd.write(js)
fd.close()


```
